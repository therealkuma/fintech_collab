```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from pathlib import Path
from hvplot import pandas

loan_data = Path("../fintech_collab/loan_data.xlsx")

loan_df = pd.read_excel(loan_data, header=0, index_col="Loan ID")

#loan_df['Purchase_Date'] = (datetime.utcfromtimestamp(0) + timedelta(loan_df['Purchase Date']).strftime("%Y-%m-%d"))

loan_df.columns = loan_df.columns.str.strip()
loan_df.columns 

#for i in loan_df.columns

b = loan_df['Beginning Period Principal Outstanding'].sum()
e = loan_df['Ending Principal Outstanding'].sum()
print(f"Beginning balance is {b:,}")
print(f"Ending balance is {e:,}")
# loan_df.columns
loan_df.head()

loan_df_pie = loan_df.groupby('Category')['Ending Principal Outstanding'].sum()
# loan_df.plot(x='Funded_Date', legend=True, stacked=True, rot=90)
loan_df_pie.plot.pie()
loan_df_pie

loan_df.head()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
from pathlib import Path
from hvplot import pandas

loan_data = Path("../fintech_collab/loan_data.xlsx")
loan_df.head()


sns.set_context("talk", font_scale= 1)
plt.figure(figsize=(25,6))
sns.scatterplot(x="Loan Amount", y="Days Past Due", data=loan_df)
plt.xlabel("Loan Amount")
plt.ylabel("Days Past Due")
plt.savefig("Seaborn_scatterplot.png",
                    format='png',dpi=150)
```
